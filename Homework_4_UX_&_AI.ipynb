```
from transformers import pipeline

# Initialize the GPT-2 text generation pipeline
generator = pipeline('text-generation', model='gpt2')

# Define the explicit prompt for the motivational speech
prompt = (
    "You are a motivational speech generator. Follow the instructions EXACTLY. "
    "Write a motivational speech with the following exact format, each section on a new line:\n\n"
    "Opening: [A strong, uplifting opening statement that sets the tone]\n"
    "Main Message: [A clear message that addresses overcoming challenges or pursuing dreams]\n"
    "Personal Anecdote: [A brief, relatable personal story or example that reinforces the main message]\n"
    "Call to Action: [A compelling conclusion that urges the listener to take positive steps]\n\n"
    "Now, generate the motivational speech exactly in this format."
)

# Generate text with the specified prompt
output = generator(prompt, max_length=300, temperature=0.75, num_return_sequences=1)
result = output[0]['generated_text']

# Optional: Print the output and check for the structure
print("Generated Motivational Speech:\n")
print(result)
```

```
prompt = (
    "You are a meditation script generator. Follow the instructions EXACTLY. "
    "Do NOT produce a story. Write a meditation script with the following exact format, each on a new line:\n\n"
    "Introduction: [A calming opening statement]\n"
    "Body Scan: [Step-by-step guide to relax each part of the body]\n"
    "Breathing Exercise: [Instructions for slow, deep breaths]\n"
    "Visualization: [A serene and vivid mental image]\n"
    "Conclusion: [Encouraging words to end the meditation]\n\n"
    "Now, generate the meditation script exactly in this format."
)

# Generate text using the ultra-explicit prompt
output = generator(prompt, max_length=300, temperature=0.7, num_return_sequences=1)
result = output[0]['generated_text']

# Post-process to try and extract our desired structure (optional)
# This is a basic check that might help if the model doesn't follow the format.
if "Introduction:" in result and "Body Scan:" in result and "Breathing Exercise:" in result and "Visualization:" in result and "Conclusion:" in result:
    print("Generated Meditation Script:\n")
    print(result)
else:
    print("The output did not follow the exact format. Try running the prompt again.\n")
    print("Output received:\n", result)
```

```
prompts = [
    (
        "Generate a meditation script for stress relief. The script MUST have the following exact sections, each on a new line, "
        "with the section header followed by its content. Do not write it as a story. Use this exact format:\n"
        "Introduction: [Your calming opening here]\n"
        "Body Scan: [Guide the listener to relax different parts of the body]\n"
        "Breathing Exercise: [Describe a slow, deep breathing exercise]\n"
        "Visualization: [Create a serene mental image]\n"
        "Conclusion: [End with encouraging words for calmness]\n"
        "\nNow, write the meditation script following the above structure exactly."
    ),
    (
        "Compose a meditation script focused on gratitude. The output MUST strictly follow this format (each on a new line):\n"
        "Introduction: [Set a calm, mindful tone]\n"
        "Mindful Awareness: [Describe noticing thoughts and feelings]\n"
        "Gratitude Reflection: [List things to be thankful for]\n"
        "Deep Breathing: [Guide deep, slow breaths]\n"
        "Closing: [End with a message of peace and contentment]\n"
        "\nNow, write the meditation script following this format exactly."
    ),
    (
        "Create a meditation script aimed at enhancing focus and clarity. Your response MUST include these sections on separate lines:\n"
        "Introduction: [Briefly set the intention for focus]\n"
        "Centering: [Guide the listener to focus their mind]\n"
        "Focus Exercise: [Provide a simple exercise to clear mental clutter]\n"
        "Visualization: [Guide the listener to visualize a clear, purposeful scene]\n"
        "Conclusion: [End with a concise, uplifting closing]\n"
        "\nNow, generate the meditation script exactly in this format."
    )
]

# Generate and print the outputs with updated prompts
for prompt in prompts:
    generated = generator(prompt, max_length=300, temperature=0.7, num_return_sequences=1)
    print(f"Prompt:\n{prompt}\n")
    print("Generated Meditation Script:\n")
    print(generated[0]['generated_text'])
    print("\n" + "-"*80 + "\n")
```
